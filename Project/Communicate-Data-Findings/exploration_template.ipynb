###   Cleaning Data

```python
courses_data = df.copy()
```

## Univariate Exploration

> Let's check the important variables distribution 

#### Checking distribution of paid and free courses 

```python
plt.figure(figsize = [10, 6]) 

sb.countplot( data=courses_data ,x='is_paid',color = base_color);
df_count = courses_data.shape[0]
paid = courses_data['is_paid'].value_counts()
# Recalculating the type_counts just to have clarity.

# get the current tick locations and labels
locs, labels = plt.xticks(rotation=45) 
plt.title("Paid - Free Courses")
# loop through each pair of locations and labels
for loc, label in zip(locs, labels):

    # get the text property for the label to get the correct count
    count = paid[label.get_text()]
    pct_string = '{:0.1f}%'.format(100*count/df_count)

    # print the annotation just below the top of the bar
    plt.text(loc, count+30, pct_string, ha = 'center', color = 'black')
```

```python
sorted_counts = courses_data['is_paid'].value_counts()
plt.figure(figsize = [10, 10]) 

plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90 ,counterclock = False,autopct=lambda x: f"{round(x, 1)} %");

```

> we can notice that from this sample that free courses is only 8.4 % from the full distribution and the rest of courses is paid courses and present 91.6 % of the distribution 

#### checking absolute Counts of each level :

```python
levels = courses_data['level'].value_counts()
levels
```

#### checking relative Counts of each level :

```python
plt.figure(figsize = [10, 10]) 

sb.countplot(data=courses_data, x='level', color=base_color, order=levels.index);
df_count = courses_data.shape[0]
# Recalculating the type_counts just to have clarity.

# get the current tick locations and labels
locs, labels = plt.xticks(rotation=45) 

# loop through each pair of locations and labels
for loc, label in zip(locs, labels):

    # get the text property for the label to get the correct count
    count = levels[label.get_text()]
    pct_string = '{:0.1f}%'.format(100*count/df_count)

    # print the annotation just below the top of the bar
    plt.text(loc, count+15, pct_string, ha = 'center', color = 'black')
```

> course level distribution clearly shows that courses that is for every one is more produced to gain more money and to get bigger number of subscribers and on the other hand beginner levelk courses comes in the second place with 34.5%
